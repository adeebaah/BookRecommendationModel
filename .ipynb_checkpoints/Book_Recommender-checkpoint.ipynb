```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
print(ratings_with_name['Book-Rating'].dtype)
```

```python
print(ratings_with_name['Book-Rating'].mean())
```

```python
print(ratings_with_name['Book-Title'].unique()[:10])  # Check a few unique titles
```

```python
test_group = ratings_with_name.groupby('Book-Title').size()
print(test_group.head())
```

```python
subset = ratings_with_name.head(100)  # Adjust the number for the subset size
avg_rating_subset = subset.groupby('Book-Title').mean()['Book-Rating'].reset_index()
print(avg_rating_subset)
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

# Group by 'Book-Title' and calculate the mean
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df.merge(books,on='Book-Title')
```
