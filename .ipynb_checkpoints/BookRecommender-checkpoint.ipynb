```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
import numpy as np
import pickle

# Load necessary artifacts
model = pickle.load(open('artifacts/model.pkl', 'rb'))
book_names = pickle.load(open('artifacts/book_names.pkl', 'rb'))
final_rating = pickle.load(open('artifacts/final_rating.pkl', 'rb'))
book_pivot = pickle.load(open('artifacts/book_pivot.pkl', 'rb'))

# Assuming you have a function to get top N recommendations for a user
def get_top_n_recommendations(book_name, N):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=N+1)
    suggested_books = [book_pivot.index[idx] for idx in suggestion.flatten()]
    suggested_books.remove(book_name)  # Remove the input book from suggestions
    return suggested_books

# Select a few users to evaluate
user_ids_to_evaluate = [237, 240, 238, 241, 184, 536]  # These are user indices from your code

precision_sum = 0
recall_sum = 0
f1_sum = 0

for user_id in user_ids_to_evaluate:
    # Assuming 'Harry Potter and the Chamber of Secrets (Book 2)' is a book the user has interacted with
    true_positive_items = ['Harry Potter and the Chamber of Secrets (Book 2)']
    
    recommended_items = get_top_n_recommendations(book_names[user_id], N=5)
    
    # Evaluate precision and recall at k
    relevant_items = set(true_positive_items)
    retrieved_items = set(recommended_items)
    
    precision = len(relevant_items.intersection(retrieved_items)) / len(retrieved_items) if len(retrieved_items) > 0 else 0
    recall = len(relevant_items.intersection(retrieved_items)) / len(relevant_items) if len(relevant_items) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    f1_sum += f1
    
    precision_sum += precision
    recall_sum += recall
    

average_precision = precision_sum / len(user_ids_to_evaluate)
average_recall = recall_sum / len(user_ids_to_evaluate)
average_f1 = f1_sum / len(user_ids_to_evaluate)

print(f"Average Precision@5: {average_precision}")
print(f"Average Recall@5: {average_recall}")
print(f"Average F1 Score@5: {average_f1:.4f}")
```
