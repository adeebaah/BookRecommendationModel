```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(final_rating, test_size=0.2, random_state=42)

# Train the model on the training data
train_book_pivot = train_data.pivot_table(columns='user_id', index='title', values='rating').fillna(0)
train_book_sparse = csr_matrix(train_book_pivot)
model = NearestNeighbors(algorithm='brute')
model.fit(train_book_sparse)

# Function to evaluate recommendations
def evaluate_recommendations(model, train_pivot, test_data):
    total_users = len(test_data['user_id'].unique())
    correct_recommendations = 0

    for user_id in test_data['user_id'].unique():
        user_test_data = test_data[test_data['user_id'] == user_id]
        for idx, row in user_test_data.iterrows():
            book_name = row['title']
            matching_books = [b for b in train_pivot.index if book_name.lower() in b.lower()]

            if len(matching_books) > 0:
                # Take the first matching book
                book_id = train_pivot.index.get_loc(matching_books[0])
                distance, suggestion = model.kneighbors(train_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

                recommended_books = train_pivot.index[suggestion[0]]
                if book_name in recommended_books:
                    correct_recommendations += 1

    accuracy = (correct_recommendations / len(test_data)) * 100
    print(f"Accuracy: {accuracy:.2f}%")

# Evaluate the recommendations on the test data
evaluate_recommendations(model, train_book_pivot, test_data)
```
